```python
#Load relevant libraries
import bar_chart_race as bcr
import ffmpeg
import pandas as pd
```

```python
#Read data into a Pandas dataframe
data = pd.read_excel('/users/solomonmuthui/Desktop/KE Economic Surveys/stats_clean.xlsx', sheet_name=1, index_col=0)
data
# pd.options.display.float_format = '{:.2}%'.format
```

```python
#Define column with the time variable and save in a pd Series
year_data = data.loc[:, data.columns=='Year']

#Define other columns containing economic sectors and save in a pd Series
other_data = round(data.loc[:, data.columns!='Year']*100, 2)

#Concatenate both pd Series
data = pd.concat([year_data, other_data], axis=1)
data
```

```python
#Initialize a bar chart race with the relevant parameters and save in a dir
bcr.bar_chart_race(df=data, title="Sector Contribution to Kenya's GDP (1955-2021) in %", 
                   period_length = 540, period_fmt = '{x:10.0f}', bar_size = 0.8, 
                   cmap = 'tab20', figsize= (6, 4), 
                   filename='/users/solomonmuthui/Desktop/KE Economic Surveys/econ_stats.mp4')
```
